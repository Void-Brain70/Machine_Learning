```python
# dataset

import os
import io
import pandas as pd
import matplotlib.pyplot as plt # for ploting
import seaborn as sns # for ploting 3D data
%matplotlib inline

df = pd.read_csv('/content/petrol_consumption.csv')
df.head()
```

```python
df = df.rename(columns={'Petrol_tax': 'x1', 'Average_income': 'x2', 'Paved_Highways': 'x3', 'Population_Driver_licence(%)': 'x4', 'Petrol_Consumption': 'y'})
df.head()
```

```python
x1 = df.x1
x2 = df.x2 
x3 = df.x3 
x4 = df.x4
y  = df.y 
```

```python
# parameter initialization
theta = [1,1.5,2,1.7,2.2]

print(theta)
```

```python
J = 30
theta = [1,1.5,2,1.7,2.2]
for i in range(30):
    # hypothesis function
    
    h = []
    m = len(y)
    for i in range(m):
        h.append(theta[0] + theta[1]*x1[i] + theta[2]*x2[i] + theta[3]*x3[i] + theta[4]*x4[i])

    print("Hypothesis Function Value is: ", h)

    # cost

    error = 0
    for i in range(m):
        error = error + (h[i] - y[i]) ** 2

    J = (1/ (2*m))*error

    print("Cost function value is", J)


    # Gradient Descent
    sum0 = 0
    sum1 = 0
    sum2 = 0
    sum3 = 0
    sum4 = 0
    for i in range(m):
        sum0 = sum0 + (h[i] - y[i])
        sum1 = sum1 + (h[i] - y[1])*x1[i]
        sum2 = sum2 + (h[i] - y[i])*x2[i]
        sum3 = sum3 + (h[i] - y[i])*x3[i]
        sum4 = sum4 + (h[i] - y[i])*x4[i]


    alpha = 0.001
    theta[0] = theta[0] - (alpha/m)*sum0
    theta[1] = theta[1] - (alpha/m)*sum1
    theta[2] = theta[2] - (alpha/m)*sum2
    theta[3] = theta[3] - (alpha/m)*sum3
    theta[4] = theta[4] - (alpha/m)*sum4

    print("Updated parameters are: ", theta)
```
